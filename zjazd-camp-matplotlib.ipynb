```python
t = np.linspace(0, 2*np.pi, 200) # okrąg
y = np.sin(t)
x = np.cos(t)

plt.fill(x,y, linewidth=8, edgecolor = "k")
plt.plot(0,0,"ko")
plt.axis("equal")
plt.text(0,0,f"(0,0)")
#plt.xlabel()
#plt.ylabel()
plt.show()
```

```python
x0 = 4
y0 = 3
r = 2
def kolo(x0, y0, r, n=200):
    t = np.linspace(0, 2*np.pi, n)
    x = x0 + r * np.cos(t)
    y = y0 + r * np.sin(t)
    plt.fill(x,y, linewidth=8, edgecolor="r",
            facecolor="#FF00FF55")
    plt.plot(x0, y0, "ko")
    plt.axis("equal")
    plt.text(x0,y0,f"({x0},{y0})",
             horizontalalignment = "left",
             verticalalignment = "top")
# plt.axis([-1.5, 3, -4, 4])
kolo(4,3,3)
kolo(2,5,1)
kolo(-5,-3,1)
plt.show()
```
